Installing Packages:

```python
import iDEA as idea
import numpy as np
import matplotlib.pyplot as plt
```

initialising a predefined atom system with two electrons and two protons:

```python
atom=idea.system.systems.atom
print(atom)
```

Plotting the atom's external potential v_ext(x):

```python
plt.plot(atom.x, atom.v_ext, 'm-')
plt.ylabel("V_ext")
plt.xlabel("x")
plt.show()
```

finding the exact ground state solution of the electrons:

```python
ground_state = idea.methods.interacting.solve(atom, k=0)
```

```python
idea.state.save_many_body_state(ground_state,"test")
```

```python
print(ground_state.space.shape)
print(ground_state.spin.shape)
print(ground_state.full.shape)
```

calculating charge density observable

```python
n = idea.observables.density(atom, state=ground_state)

plt.plot(atom.x, atom.v_ext, "m-", label="v_ext(x)")
plt.plot(atom.x, n, "k-", label="n(x)")
plt.legend()
plt.show()
```

computing the up and down spin densities seperately:

```python
n, up_n, down_n = idea.observables.density(atom, state=ground_state, return_spins=True)
plt.plot(atom.x, n, 'k-', label='n(x)')
plt.plot(atom.x, up_n, 'r-', label='n_up(x)')
plt.plot(atom.x, down_n, 'g--', label='n_down(x)')
plt.xlim([-5.0, 5.0])
plt.legend()
plt.show()
```

defining a system, s, whihc contains two electrons (both spin up) in a QHO.

```python
#define an x grid from -10 to 10 with 150 points:
x = np.linspace(-10,10,150)
#define external potential, angular freq = 0.25
v_ext = 0.5*(0.25**2)*(x**2)
#choose interaction, for this choose softened.
v_int = idea.interactions.softened_interaction(x)
```

Build the system:

```python
s = idea.system.System(x,v_ext,v_int,electrons="uu")
s2 = idea.system.System(x,v_ext,v_int,electrons="ud")
print(s)
print(f"number of electrions = {s.count}")
```

Plot external potential:

```python
plt.plot(s.x, s.v_ext, "m-")
plt.xlabel("x")
plt.ylabel("v_ext")
plt.show()
```

plotting the interaction between the electrons:

```python
plt.imshow(s.v_int)
plt.xlabel("x")
plt.ylabel("x'")
plt.show()
```

Solving the schrodinger equation for this system:

```python
ground_state = idea.methods.interacting.solve(s, k=0)
ground_state2 = idea.methods.interacting.solve(s2, k=0)
print(ground_state.space.shape)
print(ground_state.spin.shape)
print(ground_state.full.shape)
print(ground_state2.space.shape)
print(ground_state2.spin.shape)
print(ground_state2.full.shape)
```

plot spacial part of the wavefunction:

```python
plt.imshow(ground_state.space.real, cmap="seismic", vmax=np.max(ground_state.space.real), vmin=-np.max(ground_state.space.real))
plt.title("1D wavefunction for two up spin electrons in a QHM")
plt.xlabel("x")
plt.ylabel("x'")
plt.show()
plt.imshow(ground_state2.space.real, cmap="seismic", vmax=np.max(ground_state2.space.real), vmin=-np.max(ground_state2.space.real))
plt.title("1D wavefunction for one up and one down spin electrons in a QHM")
plt.xlabel("x")
plt.ylabel("x'")
plt.show()
```

```python
#test that wavefunctions are antisymmertrised
print(f"uu wavefunction: {np.allclose(ground_state.full.real[80,0,50,0], -ground_state.full.real[50,0,80,0])}")
print(f"ud wavefunction: {np.allclose(ground_state2.full.real[80,0,50,0], -ground_state2.full.real[50,0,80,0])}")
```

```python
print(f"uu energy: {ground_state.energy}")
print(f"ud energy: {ground_state2.energy}")
```

Solve for excited states:

```python
first_state = idea.methods.interacting.solve(s, k=1)
first_state2= idea.methods.interacting.solve(s2, k=1)
```

```python
plt.imshow(first_state.space.real, cmap="seismic", vmax=np.max(first_state.space.real), vmin=-np.max(first_state.space.real))
plt.title("1D wavefunction for two up spin electrons in a QHM. First excited state")
plt.xlabel("x")
plt.ylabel("x'")
plt.show()

plt.imshow(first_state2.space.real, cmap="seismic", vmax=np.max(first_state2.space.real), vmin=-np.max(first_state2.space.real))
plt.title("1D wavefunction for one up and one down spin electrons in a QHM. First exited state")
plt.xlabel("x")
plt.ylabel("x'")
plt.show()
```

Computing the optical gap:

```python
print ("uu optical gap =", first_state.energy - ground_state.energy)
print ("ud optical gap =", first_state2.energy - ground_state2.energy)
```

Computing and visualising observables:

```python
n = idea.observables.density(s, state=ground_state)
v_h = idea.observables.hartree_potential(s, n)
E_h = idea.observables.hartree_energy(s, n, v_h)


p = idea.observables.density_matrix(s, state=ground_state)
v_x = idea.observables.exchange_potential(s, p)
E_x = idea.observables.exchange_energy(s, p, v_x)


print("uu hartree energy =", E_h, ", exchange energy =", E_x)

n2 = idea.observables.density(s2, state=ground_state2)
v_h2 = idea.observables.hartree_potential(s2, n2)
E_h2 = idea.observables.hartree_energy(s2, n2, v_h2)


p2 = idea.observables.density_matrix(s2, state=ground_state2)
v_x2 = idea.observables.exchange_potential(s2, p2)
E_x2 = idea.observables.exchange_energy(s2, p2, v_x2)


print("ud hartree energy =", E_h2, ", exchange energy =", E_x2)

plt.plot(s.x, n, 'k-', label="density")
plt.plot(s.x, s.v_ext, 'm-', label="external potential")
plt.plot(s.x, v_h, 'g-', label="hartree potential")
plt.legend()
plt.xlabel("x")
plt.title("Ground state QHO uu electrons")
plt.show()

plt.plot(s2.x, n2, 'k-', label="density")
plt.plot(s2.x, s2.v_ext, 'm-', label="external potential")
plt.plot(s2.x, v_h2, 'g-', label="hartree potential")
plt.legend()
plt.xlabel("x")
plt.title("Ground state QHO ud electrons")
plt.show()


plt.imshow(p, cmap="seismic", vmax=np.max(p), vmin=-np.max(p))
plt.title("Density matrix of ground-state QHO uu electrons")
plt.show()

plt.imshow(p2, cmap="seismic", vmax=np.max(p2), vmin=-np.max(p2))
plt.title("Density matrix of ground-state QHO ud electrons")
plt.show()

plt.imshow(v_x, cmap="seismic", vmax=np.max(v_x), vmin=-np.max(v_x))
plt.title("Exchange potential of ground-state QHO uu electrons")
plt.show()

plt.imshow(v_x2, cmap="seismic", vmax=np.max(v_x2), vmin=-np.max(v_x2))
plt.title("Exchange potential of ground-state QHO ud electrons")
plt.show()
```

Time Dependence

```python
#define time grid
t = np.linspace(0,50,500)
#choose oscillating linear electric field:
v_ptrb = np.zeros(shape=t.shape+x.shape)
for j, ti in enumerate(t):
    v_ptrb[j,:] = -0.1 * x * np.sin(0.5*ti)
```

Animate the field:

```python
from matplotlib.animation import FuncAnimation
from IPython.display import HTML
fig = plt.figure()
ax = plt.axes(xlim=(s.x[0], s.x[-1]), ylim=(-3.0, 3.0))
line1, = ax.plot([], [], lw=2, c='m')
line2, = ax.plot([], [], lw=2, c='r', ls='--')
def init():
    line1.set_data([], [])
    line2.set_data([], [])
    return line1, line2
def animate(i):
    line1.set_data(s.x, v_ptrb[i,:])
    line2.set_data(s.x, v_ptrb[i,:])
    return line1, line2
anim = FuncAnimation(fig, animate, init_func=init, frames=125, interval=10, blit=True)
HTML(anim.to_html5_video())
```

Propagate ground state

```python
evolution = idea.methods.interacting.propagate(s, ground_state, v_ptrb, t)
```

Compute dynamic observables

```python

n = idea.observables.density(s, evolution=evolution)
v_h = idea.observables.hartree_potential(s, n)
E_h = idea.observables.hartree_energy(s, n, v_h)


p = idea.observables.density_matrix(s, evolution=evolution)
v_x = idea.observables.exchange_potential(s, p)
E_x = idea.observables.exchange_energy(s, p, v_x)

fig = plt.figure()
ax = plt.axes(xlim=(s.x[0], s.x[-1]), ylim=(-3.0, 3.0))
line1, = ax.plot([], [], lw=2, c='m', label="driving potential")
line2, = ax.plot([], [], lw=2, c='k', label="density")
line3, = ax.plot([], [], lw=2, c='g', label="hartree potential")
def init():
    line1.set_data([], [])
    line2.set_data([], [])
    line3.set_data([], [])
    return line1, line2, line3
def animate(i):
    line1.set_data(s.x, s.v_ext[:] + v_ptrb[i,:])
    line2.set_data(s.x, n[i,:])
    line3.set_data(s.x, v_h[i,:])
    return line1, line2, line3
plt.legend()
anim = FuncAnimation(fig, animate, init_func=init, frames=126, interval=10, blit=True)
HTML(anim.to_html5_video())
```

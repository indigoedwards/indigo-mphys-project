This program will use a simplied system to test the process of finding double excitations

```python
import numpy as np
import matplotlib.pyplot as plt
import iDEA as idea

#initialise parameters
x = np.linspace(-30, 30, 180)
v_int = idea.interactions.softened_interaction(x)

#set potential to square well
v_ext = np.concatenate((np.full(16,100),np.full(29,0.01),np.full(90,100),np.full(30,0),np.full(15,100)))
plt.plot(x,v_ext)
plt.show()

#setup system
system = idea.system.System(x,v_ext,v_int,electrons="uu")
```

```python
#solve for ground state
system_groundstate = idea.methods.interacting.solve(system, k=0)
system_firststate = idea.methods.interacting.solve(system, k=1)
system_secondstate = idea.methods.interacting.solve(system, k=2)
system_thirdstate = idea.methods.interacting.solve(system, k=3)
print("done")
```

```python

#plt.plot(x, idea.observables.density(system, state=system_groundstate), "m-",label="ground")
#plt.plot(x, idea.observables.density(system, state=system_firststate), "b-",label="first")
#plt.plot(x, idea.observables.density(system, state=system_secondstate), "r-",label="second")
plt.plot(x, idea.observables.density(system, state=system_thirdstate), "-",label="third")
plt.plot(x, v_ext, "g--",label = "potential")
plt.legend()
plt.ylim(-1,1)
plt.show()
```

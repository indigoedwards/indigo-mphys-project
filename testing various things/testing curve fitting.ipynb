```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
```

```python
x = np.linspace(-30,30)
noise = np.random.normal(0,0.2, len(x))
energies = np.exp(0.1*x) + noise
plt.plot(x,energies, ".")
plt.show()
```

```python
def func(x,a):
    return np.exp(a*x)
fit, test = sp.optimize.curve_fit(func, x, energies)

plt.plot(x,energies,".")
plt.plot(x,func(x,*fit))
plt.show()
```

```python
#finding energy at x=10
print(func(10,*fit))
```

```python
#making a square well potential which can move
import numpy as np
import matplotlib.pyplot as plt

distance = 20
well size = 10
x = np.linspace(-30,30,180)

v_ext = np.concatenate((np.full(np.where(x==(distance-(wellsize/2)),100),np.full(29,0.01),np.full(90,100),np.full(30,0),np.full(15,100)))

plt.plot(x,v_ext)
plt.show()
```

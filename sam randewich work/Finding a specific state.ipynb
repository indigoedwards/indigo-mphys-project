Looking for a state with 2 electrons each in their own hydrogen-like potential and each in their first excited state when "infinitely" far apart. The two wells are made ever so slightly different in depth.

```python
import numpy as np
import matplotlib.pyplot as plt
import iDEA

#help(iDEA.methods.non_interacting)

#Sets up a system of a hydrogen atom (single proton)
state_number = 1
x = np.linspace(-30,30,300)
v_ext = -0.5*np.exp(-1e-3*x**4)# -1.0 / (np.abs(x) + 1.0)
v_int = iDEA.interactions.softened_interaction(x)
s = iDEA.system.System(x, v_ext, v_int, electrons='u')

#Solves for the states
state_n = iDEA.methods.non_interacting.solve(s, k=state_number)
#Calculating the density of the state.
n = iDEA.observables.density(s, state=state_n)
#Plots the results.
plt.plot(s.x, n, 'r-')
plt.plot(s.x, v_ext, 'b-')
plt.xlabel("x")
plt.ylabel("Density")
plt.title("Density of state %s" %state_number)
plt.show()

energy_target = iDEA.methods.non_interacting.total_energy(s, state_n)
#print(state_n.energy)
print(energy_target)
```

As can be seen above, the energy becomes more negative when we bring two states together, so the required state must have an energy below -0.358..., this is searched for below.
